```python
import schedule 
import time 
import requests 
import csv 
import json 
from cloudscraper import create_scraper 
from apscheduler.schedulers.blocking import BlockingScheduler

# define a function that requests data and writes to a CSV file
def request_and_write(scheduler):
    # schedule the next call first
    scheduler.enter(60 * 60, 1, request_and_write, (scheduler,))
    # selected stations from solution 2
    stations = [[35.7048637 , 51.40965091],
    [35.70020574, 51.38115884],
    [35.69378492, 51.32364525]]
    
    for i in range(len(stations)):
        for j in range(len(stations)):
            if(i!=j):
    # set the parameters for the request
                origin_point = stations[i]
                dest_point = stations[j]
                origin_lat = origin_point[0]
                origin_lng = origin_point[1]
                dest_lat = dest_point[0]
                dest_lng = dest_point[1]

                scraper = create_scraper()

                url = "https://app.snapp.taxi/api/api-base/v2/passenger/newprice/s/6/0"
                headers = my_headers

                data = {
                  "points": [
                    {
                      "lat": origin_lat,
                      "lng": origin_lng
                    },
                    {
                      "lat": dest_lat,
                      "lng": dest_lng
                    },
                    None
                  ],
                  "waiting": None,
                  "round_trip": False,
                  "voucher_code": None,
                  "service_types": [1],
                  "hurry_price": 0,
                  "hurry_flag": None,
                  "priceriderecom": False,
                  "tag": 0
                }

                # make the request using the scraper object and get the response as JSON
                response = scraper.post(url, headers=headers, json=data)
                response_json = json.loads(response.text)
            #     # get the price from the response
                price = response_json['data']['prices'][0]['final']
                
                # get the current time in ISO format
                request_time = time.strftime("%Y-%m-%dT%H:%M:%S")
                print(request_time)

                # open the CSV file in append mode
                with open('./journey_price3.csv', 'a', newline='') as csvfile:
                    # create a CSV writer object
                    writer = csv.writer(csvfile)

                    # write the price to the CSV file
                    writer.writerow([request_time, origin_lat, origin_lng, dest_lat, dest_lng, price])
```

```python
import sched, time

# create a scheduler object
scheduler = sched.scheduler(time.time, time.sleep)

# schedule the first call
scheduler.enter(1*60, 1, request_and_write, (scheduler,))

# run the scheduler
scheduler.run()
```
