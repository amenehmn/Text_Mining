```python
from selenium import webdriver
```

```python
options = webdriver.ChromeOptions()
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option('useAutomationExtension', False)
options.add_argument("--disable-blink-features=AutomationControlled")

# -*- coding: UTF-8 -*-

driver = webdriver.Chrome(executable_path=r"F:\chromedriver_win32\chromedriver.exe", options= options)
```

```python
def pagelink():
    ad_list = []
#     for i in range(1):
 
    driver.get('https://divar.ir/s/isfahan')
    ads_len = len(driver.find_elements_by_css_selector('#app > div.kt-col-md-12-d59e3.browse-c7458 > main > div > div > div > div'))
    print(ads_len)
    for j in range(ads_len):
        try:
            link = driver.find_elements_by_css_selector('#app > div.kt-col-md-12-d59e3.browse-c7458 > main > div > div > div > div')[j].find_element_by_css_selector("a").get_attribute('href')
            ad_list.append(link)
        except:
            pass
    return ad_list

ads = pagelink()
ads
```

```python
def GetAdsFeature(url):

        Page = driver.get(url)

        Category = driver.find_element_by_class_name('kt-breadcrumbs__link').text
        
        Name = driver.find_element_by_class_name('kt-page-title__title.kt-page-title__title--responsive-sized').text.replace('\n', ' ')
        
        Discription = driver.find_element_by_class_name("kt-description-row__text.kt-description-row__text--primary").text.replace('\n', ' ')
        try:   
            Price = driver.find_element_by_xpath("//*[text()='قیمت']//parent::div//following-sibling::div").find_element_by_class_name('kt-unexpandable-row__value').text
        except:
#             Price = driver.find_element_by_xpath("//*[text()='قیمت کل']//parent::div//following-sibling::div").find_element_by_class_name('kt-unexpandable-row__value').text
            Price = '--'

        
#         phone_link = driver.find_element_by_xpath("//*[text()='اطلاعات تماس']//parent::button").click()
        
         
        return Category,Name,Discription,Price,url

Test = GetAdsFeature(ads[12])

Test
```

```python
from pandas import DataFrame

adsfinallistt = []
counter  = 0
len_ads = len(ads)
for i in ads:
    try:
        x = GetAdsFeature(i)
        
        Category= x[0]
        Name=x[1]
        Discription=x[2]
        Price=x[3]
        url=x[4]
        
        counter += 1
        print(counter,"/",len_ads)
        
        adsfinallistt.append([Category,Name,Discription,Price,url])
        
    except Exception as e:
        counter += 1
        print(counter,"/",len_ads)
        print(e)

df = DataFrame(adsfinallistt,columns=['Category','Name','Discription','Price','url'])


df.to_csv(r"./exportFinal_divar.csv", sep='|', encoding='utf-8')

df
```
