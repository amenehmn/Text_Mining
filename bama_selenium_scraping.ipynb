```python
from selenium import webdriver
```

```python
options = webdriver.ChromeOptions()
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option('useAutomationExtension', False)
options.add_argument("--disable-blink-features=AutomationControlled")

driver = webdriver.Chrome(executable_path=r"F:\chromedriver_win32\chromedriver.exe", options= options)
```

```python
carlist = []
# Navigate to a web page
driver.get("https://bama.ir/car")

# Define the desired number of div elements
desired_number = 50
previous_number = 0

# Find all the div elements on the page
#__layout > div > div > section > div.bama-adlist-content-holder > div > div:nth-child(1) > a
div_elements = driver.find_elements_by_css_selector("#__layout > div > div > section > div.bama-adlist-content-holder > div > div")

# Get the current number of div elements
current_number = len(div_elements)
print(current_number)

for j in range(current_number):
    try:
        link = driver.find_elements_by_css_selector( "#__layout > div > div > section > div.bama-adlist-content-holder > div > div")[j].find_element_by_css_selector("a").get_attribute('href')
        carlist.append(link)
    except:
        pass

# While the current number is less than the desired number
while current_number < desired_number:
  # Scroll down by 100 pixels
    carlist = []
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
  
  # Find all the div elements on the page again
    div_elements = driver.find_elements_by_css_selector( "#__layout > div > div > section > div.bama-adlist-content-holder > div > div")

  # Get the new number of div elements
    current_number = len(div_elements)
    print(current_number)
    for j in range(current_number):
        try:
            link = driver.find_elements_by_css_selector("#__layout > div > div > section > div.bama-adlist-content-holder > div > div")[j].find_element_by_css_selector("a").get_attribute('href')
            carlist.append(link)
        except:
            pass
  
  # Break the loop if there are no more div elements to load
    if current_number == previous_number:
        break
  
  # Store the previous number of div elements
    previous_number = current_number

# Do something with the div elements
print(carlist)
```

```python
import time

def GetCarFeature(url):
    try:
        page = driver.get(url)

        Name = driver.find_element_by_class_name('bama-ad-detail-title__title').text
        
        Kilometer = driver.find_element_by_css_selector('#__layout > div > div > div > div > div.info-wrapper > div.bama-vehicle-detail-with-icon > div:nth-child(1) > span').text

        Year = driver.find_element_by_class_name("bama-ad-detail-title__subtitle").text[:4]
#         except:
#             Year = driver.find_element_by_css_selector('#content > div.prd-detail.ad-detail-maindiv > div.prdright > div.prdinfo > div.inforow.newinfo > div.inforight > div > div > h1 > span:nth-child(5)').text


        Type = driver.find_element_by_css_selector('#__layout > div > div > div > div > div.info-wrapper > div.bama-vehicle-detail-with-icon > div:nth-child(3) > span').text

        
        Sokht = driver.find_element_by_css_selector('#__layout > div > div > div > div > div.info-wrapper > div.bama-vehicle-detail-with-icon > div:nth-child(2) > span').text

        
        State = driver.find_element_by_class_name('address-text').text

#         City = driver.find_element_by_id('#content > div.prd-detail.ad-detail-maindiv > div.prdright > div.prdinfo > div.inforow.newinfo > div.inforight > p:nth-child(10) > span:nth-child(2)').text

        Color = driver.find_element_by_css_selector('#__layout > div > div > div > div > div.info-wrapper > div.bama-vehicle-detail-with-icon > div:nth-child(4) > span').text
#         except:
#             Color = driver.find_element_by_css_selector('#content > div.prd-detail.ad-detail-maindiv > div.prdright > div.prdinfo > div.inforow.newinfo > div.inforight > p:nth-child(11) > span:nth-child(2)').text

        try:
            Desc = driver.find_element_by_css_selector('#__layout > div > div > div > div > div.description-wrapper > div.description > div > div.desc > p').text
        except:
            Desc = ''

        
        Price =  driver.find_element_by_class_name('bama-ad-detail-price__price-text').text


#         Phone_open = driver.find_element_by_link_text("[نمایش کامل]").click()
        Phone_open = driver.find_element_by_css_selector("#__layout > div > div > div > div > div.info-wrapper > div.bama-call-to-seller > button > span").click()
        time.sleep(2)
        phone_list_len = len(driver.find_elements_by_css_selector("#__layout > div > div > div > div > div.info-wrapper > div.bama-call-to-seller.bama-call-to-seller--open > div.bama-call-to-seller__numbers-holder > div.bama-call-to-seller__numbers > a"))
        phone_list = []
        for i in range(phone_list_len):
            try:
                phone = driver.find_elements_by_css_selector("#__layout > div > div > div > div > div.info-wrapper > div.bama-call-to-seller.bama-call-to-seller--open > div.bama-call-to-seller__numbers-holder > div.bama-call-to-seller__numbers > a")[i].get_attribute('href')
                phone_list.append(phone)
            except:
                pass
        PhoneTotal = '-'.join(str(e) for e in phone_list)

        
#         return NameTotal,KilometerTotal.strip(),YearTotal.strip(),TypeTotal.strip(),SokhtTotal.strip(),StateTotal.strip(),CityTotal.strip(),ColotTotal.strip(),DescTotal.strip(),PriceTotal,url 
        return Name,Kilometer.strip(),Year.strip(),Type.strip(),Sokht.strip(),State.strip(),Color.strip(),Desc.strip().replace("\u200c","").replace("/",','),Price,PhoneTotal,url 
    except Exception as e:
        print(e)

Test = GetCarFeature(carlist[4])

Test
```

```python
from pandas import DataFrame
# Import the progressbar module
import progressbar

carfinallistt = []
counter  = 0
lencars = len(carlist)
# Create a progress bar with a simple format
bar = progressbar.ProgressBar(max_value=lencars, widgets=[progressbar.Bar(), ' ', progressbar.Percentage(), ' ', progressbar.ETA()])
for car in carlist:
    try:
        x = GetCarFeature(car)
        NameTotal= x[0]
        KilometerTotal=x[1]
        YearTotal=x[2]
        TypeTotal=x[3]
        SokhtTotal=x[4]
        StateTotal=x[5]
#         CityTotal=x[6]
        ColotTotal=x[6]
        DescTotal=x[7]
        PriceTotal=x[8]
        PhoneTotal = x[9]
        url=x[10]
        bar.update(counter + 1)
        counter += 1
        carfinallistt.append([NameTotal,KilometerTotal,YearTotal,TypeTotal,SokhtTotal,StateTotal,ColotTotal,DescTotal,PriceTotal,PhoneTotal,url])
    except Exception as e:
        bar.update(counter + 1)
        counter += 1
        print(counter,"/",lencars)
        print(e)

df = DataFrame(carfinallistt,columns=['Name','Kilometer','Year','Type','Sokht','State','Color','Desc','Price','PhoneTotal','url'])


df.to_csv(r"./exportFinal_selenium.csv", sep='|', encoding='utf-8')

df
```
